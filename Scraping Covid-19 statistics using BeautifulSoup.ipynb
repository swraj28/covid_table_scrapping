```python
import requests
from bs4 import BeautifulSoup as soup
```

```python
url="https://www.worldometers.info/coronavirus/countries-where-coronavirus-has-spread/"
```

```python
r=requests.get(url)
print(r.content)
```

```python
url_data=soup(r.text,'html.parser')
print(url_data)
```

```python
url_data.findAll('h1',{})
```

```python
tables=url_data.findAll('table')
```

```python
print(len(tables))
print(type(tables))
print(tables[0])
```

```python
tables=url_data.findAll('table',{'class':'table table-striped table-bordered','id':'table3'})
```

```python
print(len(tables))
covid_table=tables[0]
```

```python
print(covid_table)
```

```python
headers=covid_table.findAll('th')
print(len(headers))
print(headers)
```

```python
print(headers[0].text)
```

```python
# We will use these headers as columns names hence we have to store them in a variable
column_titles = [ct.text for ct in headers]
# slice out the carriage return (\n)
# column_titles = [ct.text[:-1] for ct in headers]
print(column_titles)
```

```python
rows_data = covid_table.findAll('tr')[1:]
print("Total number of rows: {}".format(len(rows_data)))
```

```python
table_rows = []
for row_id, row in enumerate(rows_data):
    
    # We will skip the last row as it contains missing value
    current_row = []
    row_data = row.findAll('td', {})
    
    for col_id, data in enumerate(row_data):   
        # We will also replace any commas(',') present in the data as we have to store the data in a csv
        # file
        text = data.text
        text = text.replace(",", "") # We are slicing the data as all the elemnets contain a carriage
        # return('\n') at the end
        current_row.append(text)
            
    table_rows.append(current_row)
```

```python
print(table_rows)
```

```python
filename = 'covid_stats.csv'
with open(filename,'w',encoding='utf-8') as f:
    #Write the header
    header_string = ','.join(column_titles)
    header_string += '\n'
    f.write(header_string)
    
    for row in table_rows[:-1]:
        row_string = ""
        for w in row:
            w =  w.replace(',','')
            row_string += w +','
        row_string = row_string[:-1]    
        row_string += '\n'
        f.write(row_string)
```
